```python
import glob
import matplotlib.pyplot as plt
import random
import matplotlib.image as mpimg
%matplotlib inline

def plot_images(img1, img2, title1, title2, colormap1, colormap2):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2))
    f.tight_layout()
    ax1.imshow(img1, cmap=colormap1)
    ax1.set_title(title1, fontsize=10)
    ax2.imshow(img2, cmap=colormap2)
    ax2.set_title(title2, fontsize=10)
    ax1.axis('off');
    ax2.axis('off');
    f.savefig("output_images/" + title2.replace(" ",""), bbox_inches='tight', transparent=True)
```

# Load Files

```python
# Load vehicle and non-vehicle examples
cars = glob.glob('vehicles\*\*.png')
notcars = glob.glob('non-vehicles\*\*.png')    
        
print('Vehicle examples: {}'.format(len(cars)))
print('Non-vehicle examples: {}'.format(len(notcars)))

# Plot the examples
car_example = mpimg.imread(cars[random.randint(0, len(cars))])
notcar_example = mpimg.imread(notcars[random.randint(0, len(notcars))])
plot_images(car_example, notcar_example, "Car example", "Not car example", 'jet', 'jet')
```
